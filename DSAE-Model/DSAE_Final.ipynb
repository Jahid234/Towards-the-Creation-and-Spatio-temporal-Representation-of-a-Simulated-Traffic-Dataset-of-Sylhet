```python
inp = np.array(df).flatten('C')
wrappedinput = np.array([output])
wrappedoutput = np.array([inp])


inp1 = np.array(df1).flatten('C')
wrappedinput1 = np.array([output1])
wrappedoutput1 = np.array([inp1])


inp2 = np.array(df2).flatten('C')
wrappedinput2 = np.array([output2])
wrappedoutput2 = np.array([inp2])


inp3 = np.array(df3).flatten('C')
wrappedinput3 = np.array([output3])
wrappedoutput3 = np.array([inp3])


inp4 = np.array(df4).flatten('C')
wrappedinput4 = np.array([output4])
wrappedoutput4 = np.array([inp4])


inp5 = np.array(df5).flatten('C')
wrappedinput5 = np.array([output5])
wrappedoutput5 = np.array([inp5])


inp6 = np.array(df6).flatten('C')
wrappedinput6 = np.array([output6])
wrappedoutput6 = np.array([inp6])


inp7 = np.array(df7).flatten('C')
wrappedinput7 = np.array([output7])
wrappedoutput7 = np.array([inp7])


inp8 = np.array(df8).flatten('C')
wrappedinput8 = np.array([output8])
wrappedoutput8 = np.array([inp8])

print(wrappedinput)
```

```python
#Create model

model = Sequential()
model.add(Dense(2000, input_dim=wrappedinput.shape[1], activation='relu'))
model.add(Dense(1000, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(1095))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(wrappedinput,wrappedoutput,verbose=0,batch_size=len(wrappedinput),epochs=10000)

pred = model.predict(wrappedinput)

print("Actual")
print(wrappedoutput)

print("Pred")
print(pred)
```
